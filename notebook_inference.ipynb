### **Model Inference**

Import Library

```python
import pandas as pd
import numpy as np
import joblib as joblib
```

Load Files

```python
scaler = joblib.load('deployment/scaler.pkl')
hot = joblib.load('deployment/hot_encoder.pkl')
ord = joblib.load('deployment/ord_encoder.pkl')

#model
rf = joblib.load('deployment/rf_model.pkl')
```

Data Inf using individual files

```python
data_inf = {'model': {362: 'X5',
                    4314: 'Q5',
                    7862: 'Q3',
                    9473: '4 Series',
                    4363: 'Tucson'},
            'year': {362: 2020, 4314: 2019, 7862: 2017, 9473: 2015, 4363: 2017},
            'price': {362: 60990, 4314: 44495, 7862: 20000, 9473: 14299, 4363: 14270},
            'transmission': {362: 'Semi-Auto',
                            4314: 'Semi-Auto',
                            7862: 'Automatic',
                            9473: 'Automatic',
                            4363: 'Manual'},
            'mileage': {362: 3500, 4314: 8700, 7862: 21981, 9473: 80614, 4363: 14752},
            'fuelType': {362: 'Diesel',
                        4314: 'Hybrid',
                        7862: 'Petrol',
                        9473: 'Diesel',
                        4363: 'Petrol'},
            'tax': {362: 145, 4314: 135, 7862: 145, 9473: 125, 4363: 145},
            'mpg': {362: 33.6, 4314: 117.7, 7862: 47.9, 9473: 58.9, 4363: 44.8},
            'engineSize': {362: 3.0, 4314: 2.0, 7862: 1.4, 9473: 2.0, 4363: 1.6},
            'company': {362: 'BMW',
                        4314: 'Audi',
                        7862: 'Audi',
                        9473: 'BMW',
                        4363: 'Hyundai'}}
```

```python
data_inf=pd.DataFrame(data_inf).reset_index(drop=True)
data_inf
```

```python
# seleksi kolom
num_col = ['mileage', 'engineSize','mpg']
cat_ord = ['year','transmission']
cat_nom = ['fuelType','model','company']

# transform dari file
data_scaled = scaler.transform(data_inf[num_col])
data_hot = hot.transform(data_inf[cat_nom]).toarray()
data_ord = ord.transform(data_inf[cat_ord])
```

```python
data = np.concatenate([data_scaled, data_ord, data_hot], axis=1) # gabung file
```

```python
pred = rf.predict(data) # predict 
data_inf['prediction'] = pred # masukan hasil prediksi
data_inf
```
